## Create the word frequency dictionary for each class

Create a word frequency dictionary for each class.

```python
def word_bag(text):
    bag = {}
    for word in text.split():
        bag[word] = bag.get(word, 0) + 1
    return bag

spam_words ={}
for n in range(0,len(train_spam)):
    bag = word_bag(train_spam.iloc[n]['text'])
    for word in bag:
        spam_words[word] = spam_words.get(word,0) + bag[word]

ham_words = {}
for n in range(0,len(train_ham)):
    bag = word_bag(train_ham.iloc[n]['text'])
    for word in bag:
        ham_words[word] = ham_words.get(word,0) + bag[word]
```
